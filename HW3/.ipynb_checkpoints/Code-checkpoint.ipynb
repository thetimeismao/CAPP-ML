## Step 5. Classify

```python
models = ['LR', 'KNN', 'DT', 'SVM', 'RF', 'GB']
metrics = ['precision', 'recall', 'f1', 'auc']
```

```python
# We split up our data into training and testing sets.
X_train, X_test, y_train, y_test = train_test_split(data[features], data[label])
train = pd.concat([y_train, X_train], axis = 1)

# Replace our dataframe with only the training data 
data = train
```

```python
all_models = P5_classify.classify(data[features],
                                  data[label],
                                  models, 
                                  3, 
                                  0.05, 
                                  metrics)    
```

```python
all_models
```

```python
auc = P5_classify.select_best_models(all_models, 
                                     models, 
                                     'auc')

precision = P5_classify.select_best_models(all_models, 
                                           models, 
                                           'precision')

recall = P5_classify.select_best_models(all_models, 
                                        models, 
                                        'recall')

recall = P5_classify.select_best_models(all_models, 
                                        models, 
                                        'recall')

f1 = P5_classify.select_best_models(all_models, 
                                    models, 
                                    'f1')
```
